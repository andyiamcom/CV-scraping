```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def get_CV(contents):
    data_columns = ['nr','amet', 'tööandja','asukoht', 'sisestatud','aegub','kuupalk']
    
    CV_data = pd.DataFrame(columns = data_columns, index = range(0,1))
    CV_table = pd.DataFrame(columns = data_columns)
   
    try:
        soup = bs(contents.text, 'html.parser')
        rows = soup.findAll("tr", {"class":"f_job_row"})
    except:
        print('Could not parse web contents')
        return CV_table
    
    i = 0
    for row in rows:
        
        
        try: 
            row.find("td", {"class":"column column1 faded hidden-xs-down js_left_indent"}).find \
        ("span", {"class":"f_vip"}).decompose()
        except:
            pass
        
        try: 
            row.find("td", {"class":"column column1 faded hidden-xs-down js_left_indent"}).find \
        ("span", {"class":"f_new"}).decompose()
        except:
            pass
        
        
        sisestatud = row.find("td", {"class":"column column1 faded hidden-xs-down js_left_indent"}).text.strip()
        
        print(sisestatud)
        amet = row.find("a", {"class":"f_job_title main_job_link"}).text
        tööandja = row.find("span", {"class":"f_job_company"}).text
        asukoht =  row.find("span", {"class":"f_job_city"}).text.strip()
        aegub = row.find("td", {"class":"column column5 faded hidden-xs-down text-right"}).text.splitlines()[1].strip()
        
        try:
            kuupalk = row.find("span", {"class":"f_job_salary"}).text
        except:
            kuupalk = "puudub"

        # formatting
        
        # dataframe
        CV_data['nr'] = i
        CV_data['amet'] = amet
        CV_data['tööandja'] = tööandja
        CV_data['asukoht'] = asukoht
        CV_data['sisestatud'] = sisestatud
        CV_data['aegub'] = aegub
        CV_data['kuupalk'] = kuupalk

        CV_table = CV_table.append(CV_data)
        i += 1
        
        '''
        
        
        sisestatud = re.split("\s", sisestatud)[0]
        aegub = re.split("\s", aegub)[1]
        kuupalk = kuupalk.replace('Kuupalk: ', '')
        
         
        '''
        
        
        
    message = ' Downloaded ' + str(len(CV_table)) + ' lines of CV data'
    print(message)
    
    CV_table = CV_table.set_index('nr')

    return CV_table
```

```python
link1 = 'https://www.cvkeskus.ee/joboffers.php?_track=index_click_job_search&op=search&search_location=landingpage&ga_track=homepage&search%5Bkeyword%5D=analüütik&mobile_search%5Bkeyword%5D=analüütik&tmp_city=&tmp_cat=&tmp_city=&tmp_category=&search%5Bkeyword%5D=analüütik&search%5Bexpires_days%5D=&search%5Bjob_lang%5D=&search%5Bjob_minimumpay%5D=&search%5Bjob_maximumpay%5D=&search%5Bjob_currency%5D=EUR&search%5Bjob_salary%5D=3'

contents = requests.get(link1)
first_page = get_CV(contents)
first_page.head()
```
